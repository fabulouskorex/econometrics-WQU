# 1. Data Importing

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
GLD_ETF_df = pd.read_csv("SPDR Gold Shares (GLD) Jan - Dec 2020.csv",parse_dates=True,index_col=0)
EQT_ETF_df = pd.read_csv("Amundi CAC 40 UCITS ETF-C (C40.PA) Jan 2020 - Dec 2020.csv",parse_dates=True,index_col=0)
BTC_df = pd.read_csv("Bitcoin USD (BTC-USD) Jan 2020 - Dec 2020.csv",parse_dates=True,index_col=0)
```

```python
gold_ETF_df = pd.read_csv("SPDR Gold Shares (GLD) April 2020 - December 2020.csv",parse_dates=True,index_col=0)
equity_ETF_df = pd.read_csv("Amundi CAC 40 UCITS ETF-C (C40.PA) April 2020 - December 2020.csv",parse_dates=True,index_col=0)
bitcoin_df = pd.read_csv("Bitcoin USD (BTC-USD) April 2020 - December 2020.csv",parse_dates=True,index_col=0)
```

# 2. Data Processing

```python
gold_ETF_df['Daily_return_pct'] = gold_ETF_df['Close'].pct_change(1)
gold_ETF_df.head()
```

```python
equity_ETF_df['Daily_return_pct'] = equity_ETF_df['Close'].pct_change(1)
equity_ETF_df.head()
```

```python
bitcoin_df['Daily_return_pct'] = bitcoin_df['Close'].pct_change(1)
bitcoin_df.head
```

# 3. Data Summaries 

```python
GLD_ETF_df["Close"]["2020-03-01":"2020-12-31"].plot()
moving_avg = GLD_ETF_df.rolling(20).mean()
moving_avg.tail()
```

```python
EQT_ETF_df["Close"]["2020-03-01":"2020-12-31"].plot()
moving_avg = EQT_ETF_df.rolling(20).mean()
moving_avg.tail()
```

```python
BTC_df["Close"]["2020-03-01":"2020-12-31"].plot()
moving_avg = BTC_df.rolling(20).mean()
moving_avg.tail()
```

# 4. Graphing

```python
x=GLD_ETF_df["Close"]["2020-03-01":"2020-12-31"]
moving_avg = x.rolling(20).mean()

plt.figure(figsize=(10,7))
plt.plot(x)
plt.plot(moving_avg)
plt.show()
```

```python
x=EQT_ETF_df["Close"]["2020-03-01":"2020-12-31"]
moving_avg = x.rolling(20).mean()

plt.figure(figsize=(10,5))
plt.plot(x)
plt.plot(moving_avg)
plt.show()
```

```python
x=BTC_df["Close"]["2020-03-01":"2020-12-31"]
moving_avg = x.rolling(20).mean()

plt.figure(figsize=(10,5))
plt.plot(x, label="BTC_df")
plt.plot(moving_avg, label="moving_avg")
plt.legend()
plt.show()
```

# 4.2

```python
y1_data = gold_ETF_df["Close"]
y2_data = equity_ETF_df["Close"]
fig, y1_axis = plt.subplots()
plt.figure(figsize=(10,5))

y1_color = 'tab:red'
y2_color = 'tab:blue'

y1_axis.set_xlabel('Date')
y1_axis.set_ylabel('gold_ETF_df', color=y1_color)
y1_axis.plot( y1_data, color=y1_color)
y1_axis.tick_params(axis='y', labelcolor=y1_color)
y2_axis = y1_axis.twinx()  # instantiate a second axes that shares the same x-axis


y2_axis.set_ylabel('equity_ETF_df', color=y2_color)  # we already handled the x-label with ax1
y2_axis.plot( y2_data, color=y2_color)
y2_axis.tick_params(axis='y', labelcolor=y2_color)

#fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(gold_ETF_df["Close"],label="gold_ETF_df" )
plt.plot(equity_ETF_df["Close"],label="equity_ETF_df" )
plt.legend()
plt.show()
```
